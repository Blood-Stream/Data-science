```
winner = pd.concat([df, dev_df], axis=1)
winner
```

```
winner.apply(pd.Series.value_counts)
```

```
meanshift = MeanShift().fit(winner)
```

```
print(max(meanshift.labels_))
```

```
df_pred = winner
df_pred['meanshift'] = meanshift.predict(winner)
df_pred
```

```
df_pred['meanshift'].value_counts()
```

```
mini_batch = MiniBatchKMeans(n_clusters=97, batch_size=8).fit(winner)
```

```
df_pred['minibatch'] = mini_batch.predict(winner)
df_pred
```

```
df_pred['minibatch'].value_counts()
```

```
km = KMeans(n_clusters=97).fit(winner)
```

```
df_pred['kmeans'] = km.predict(winner)
df_pred
```

```
df_pred['kmeans'].value_counts()
```

```
agg = AgglomerativeClustering(n_clusters=97).fit(winner)
```

```
df_pred['agg'] = agg.fit_predict(winner)
df_pred
```

```
df_pred['agg'].value_counts()
```

```
dbscan = DBSCAN(eps=15, min_samples=100).fit(winner)
```

```
df_pred['dbscan'] = dbscan.fit_predict(winner)
df_pred
```

```
df_pred['dbscan'].value_counts()
```

```
optics = OPTICS(min_samples=100).fit(winner)
```

```
df_pred['optics'] = optics.fit_predict(winner)
df_pred
```

```
df_pred['optics'].value_counts()
```

```
birch = Birch(n_clusters=None).fit(winner)
```

```
df_pred['birch'] = optics.fit_predict(winner)
df_pred
```

```
df_pred['birch'].value_counts()
```
